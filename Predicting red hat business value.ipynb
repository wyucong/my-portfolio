There are two files: people.csv with characteristics of each people and act_train.csv with cahracteristics of each activity. we want to identify which customer have the most potential value and worth approaching to.

```python
import pandas as pd

people = pd.read_csv('people.csv')
people.head()
```

```python
act_train = pd.read_csv('act_train.csv')
act_train.head(10)
```

We can see from the data type 1 activity has characteristis 1 to 9. and the rest of the activites have characteristic 10. Therefore, I train two different classification models.

```python
merged = act_train.merge(people,on = 'people_id')
merged.head()
```

```python
merged_type1 = merged[merged['activity_category']=='type 1']
```

```python
merged_type1.head()
```

```python
del merged_type1['char_10_x']
```

```python
def trans_to_int(series):
    item = []
    for s in series:
        item.append(s.split(' ')[1])
    return item
```

```python
del merged_type1['activity_category']
merged_type1.head()
```

```python
for i in range(3,12):
    merged_type1[merged_type1.columns[i]] = trans_to_int(merged_type1[merged_type1.columns[i]])
```

```python
for i in range(17,24):
        merged_type1[merged_type1.columns[i]] = trans_to_int(merged_type1[merged_type1.columns[i]])
merged_type1.head()
```

```python

merged_type1[merged_type1.columns[13]] = trans_to_int(merged_type1[merged_type1.columns[13]])
merged_type1[merged_type1.columns[14]] = trans_to_int(merged_type1[merged_type1.columns[14]])
merged_type1[merged_type1.columns[15]] = trans_to_int(merged_type1[merged_type1.columns[15]])
```

```python
map1 = {False: 0,
        True: 1
}
for i in range(24,(merged_type1.shape[1]-1)):
    merged_type1[merged_type1.columns[i]] = merged_type1[merged_type1.columns[i]].map(map1) 

```

```python
predictors = [c for c in merged_type1.columns if c not in ["people_id","activity_id","date_x","date_y","outcome"]]
from sklearn import cross_validation
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=10, min_weight_fraction_leaf=0.1)
scores = cross_validation.cross_val_score(clf, merged_type1[predictors], merged_type1['outcome'], cv=2)
scores
```

The scores are very good for both test sample and train samples.

```python
merged_rest = merged[merged['activity_category']!='type 1']
merged_rest.head()
```

```python
del merged_rest['char_1_x']
del merged_rest['char_2_x']
del merged_rest['char_3_x']
del merged_rest['char_4_x']
del merged_rest['char_5_x']
del merged_rest['char_6_x']
del merged_rest['char_7_x']
del merged_rest['char_8_x']
del merged_rest['char_9_x']

merged_type1.columns
```

```python
merged_rest.head()
```

```python
merged_rest[merged_rest.columns[3]] = trans_to_int(merged_rest[merged_rest.columns[3]])
merged_rest[merged_rest.columns[4]] = trans_to_int(merged_rest[merged_rest.columns[4]])
merged_rest[merged_rest.columns[6]] = trans_to_int(merged_rest[merged_rest.columns[6]])
merged_rest[merged_rest.columns[7]] = trans_to_int(merged_rest[merged_rest.columns[7]])
merged_rest[merged_rest.columns[8]] = trans_to_int(merged_rest[merged_rest.columns[8]])
```

```python

for i in range(10,17):
    merged_rest[merged_rest.columns[i]] = trans_to_int(merged_rest[merged_rest.columns[i]])
merged_rest.iloc[:,9:20].head()
```

```python

for i in range(17,(merged_rest.shape[1]-1)):
    merged_rest[merged_rest.columns[i]] = merged_rest[merged_rest.columns[i]].map(map1) 
```

```python
predictors_rest = [c for c in merged_rest.columns if c not in ["people_id","activity_id","date_x","date_y","outcome"]]
from sklearn import cross_validation
from sklearn.ensemble import RandomForestClassifier

clf2 = RandomForestClassifier(n_estimators=10, min_weight_fraction_leaf=0.1)
scores2 = cross_validation.cross_val_score(clf2, merged_rest[predictors_rest], merged_rest['outcome'], cv=2)
scores2
```
