```python
import pandas as pd
star_wars = pd.read_csv('StarWars.csv',encoding = 'ISO-8859-1')
star_wars = star_wars.drop(0)
star_wars.head()
```

```python
star_wars.columns
```

```python
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
```

```python
star_wars.head()
```

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts()
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts()
```

```python
yes_no = {
    "Yes": True,
    "No": False
}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?']= star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
star_wars.head()

```

```python
star_wars = star_wars.rename(columns = {
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4":"seen_2",
        "Unnamed: 5":"seen_3",
        "Unnamed: 6":"seen_4",
        "Unnamed: 7":"seen_5",
        "Unnamed: 8":"seen_6",
    
    })
star_wars.head()
```

```python
map1 = {
    
    "Star Wars: Episode I  The Phantom Menace": True,
    float('nan'): False
}
star_wars[star_wars.columns[3]] = star_wars[star_wars.columns[3]].map(map1)
star_wars.head()
```

```python
map2 = {
    
    "Star Wars: Episode II  Attack of the Clones": True,
    float('nan'): False
}
star_wars[star_wars.columns[4]] = star_wars[star_wars.columns[4]].map(map2)

map3 = {
    
    "Star Wars: Episode III  Revenge of the Sith": True,
    float('nan'): False
}
star_wars[star_wars.columns[5]] = star_wars[star_wars.columns[5]].map(map3)

map4 = {
    
    "Star Wars: Episode IV  A New Hope": True,
    float('nan'): False
}
star_wars[star_wars.columns[6]] = star_wars[star_wars.columns[6]].map(map4)
map5 = {
    
    "Star Wars: Episode V The Empire Strikes Back": True,
    float('nan'): False
}
star_wars[star_wars.columns[7]] = star_wars[star_wars.columns[7]].map(map5)
map6 = {
    
    "Star Wars: Episode VI Return of the Jedi": True,
    float('nan'): False
}
star_wars[star_wars.columns[8]] = star_wars[star_wars.columns[8]].map(map6)

star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars = star_wars.rename(columns = {
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10":"ranking_2",
        "Unnamed: 11":"ranking_3",
        "Unnamed: 12":"ranking_4",
        "Unnamed: 13":"ranking_5",
        "Unnamed: 14":"ranking_6",
    
    })
star_wars.head()
```

```python
rankings = star_wars[star_wars.columns[9:15]]
ranking_means = rankings.mean(axis=0)
ranking_means
```

The third one has the higher ranking (lower ranking the better)

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ind = np.arange(6)
width = 0.35
ax.bar(ind,ranking_means)
```

finally, we finished cleaning the data. it seams the first three movies have higher rankings (lower ranking the better).

```python
seens = star_wars[star_wars.columns[3:9]]
seens_sum = seens.sum(axis=0)
seens_sum
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ind = np.arange(6)
ax.bar(ind,seens_sum)
```

we can see that the last three movies are seen by more people which correspondent to the last ranking bar chart that they have better ratings.

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
male_rankings = males[males.columns[9:15]]
female_rankings = females[females.columns[9:15]]
female_ranking_means = female_rankings.mean(axis=0)
male_ranking_means = male_rankings.mean(axis = 0)
```

```python
female_ranking_means
```

```python
male_ranking_means
```

```python
fig = plt.figure(figsize = (15,6))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ind = np.arange(6)
ax1.bar(ind,female_ranking_means)
ax1.set(title = 'female rankings')
ax2.bar(ind,male_ranking_means)
ax2.set(title = 'male_rankings')
```

```python
seens_male = males[males.columns[3:9]]
seens_male_sum = seens_male.sum(axis=0)
seens_female = females[females.columns[3:9]]
seens_female_sum = seens_female.sum(axis = 0)
```

```python
seens_female_sum
```

```python
seens_male_sum
```

```python
fig = plt.figure(figsize = (15,6))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ind = np.arange(6)
ax1.bar(ind,seens_female_sum)
ax1.set(title = 'female seen')
ax2.bar(ind,seens_male_sum)
ax2.set(title = 'male seen')
```
